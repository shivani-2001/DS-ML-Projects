```python
from PIL import ImageTk, Image
root = Tk()
root.title("Hey Buddy!!!")
root.iconbitmap("E:/DE/loveaiworld.ico")

img_me = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/me.jpeg")).resize((400, 400)))
img1 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/2.jpeg")).resize((400, 400)))
img2 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/4.jpeg")).resize((400, 400)))
img3 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/6.jpeg")).resize((400, 400)))
img4 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/7.jpeg")).resize((400, 400)))
img5 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/10.jpeg")).resize((400, 400)))
img6 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/11.jpeg")).resize((400, 400)))
img7 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/12.jpeg")).resize((400, 400)))
img8 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/16.jpeg")).resize((400, 400)))
img9 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/17.jpeg")).resize((400, 400)))
img10 = ImageTk.PhotoImage((Image.open("E:/Courses/Face Recognition/FF/unknown/18.jpeg")).resize((400, 400)))

label_img = Label(image=img_me)
label_img.grid(row=0, column=0, columnspan=3)

img_list = [img_me, img1, img2, img3, img4, img5, img6, img7, img8, img9, img10]

status = Label(root, text="Image 1 of " +str(len(img_list)), bd=1, relief=SUNKEN, anchor=E)
status.grid(row=2, column=0, columnspan=3, sticky=W+E)

def forward(img_no):
    global label_img
    global button_bck
    global button_fwd
    
    label_img.grid_forget()
    label_img = Label(image=img_list[img_no-1])
    
    button_fwd = Button(root, text=">>", command=lambda: forward(img_no+1))
    button_bck = Button(root, text="<<", command=lambda: back(img_no-1))
      
    if img_no == len(img_list):
        button_fwd = Button(root, text=">>", state=DISABLED)
    
    label_img.grid(row=0, column=0, columnspan=3)    
    button_bck.grid(row=1, column=0)
    button_fwd.grid(row=1, column=2)
    
    status = Label(root, text="Image "+ str(img_no) +" of " +str(len(img_list)), bd=1, relief=SUNKEN, anchor=E)
    status.grid(row=2, column=0, columnspan=3, sticky=W+E)

def back(img_no):
    global label_img
    global button_bck
    global button_fwd
    
    label_img.grid_forget()
    label_img = Label(image=img_list[img_no-1])
    
    button_fwd = Button(root, text=">>", command=lambda: forward(img_no+1))
    button_bck = Button(root, text="<<", command=lambda: back(img_no-1))
      
    if img_no == 1:
        button_bck = Button(root, text=">>", state=DISABLED)
    
    label_img.grid(row=0, column=0, columnspan=3)    
    button_bck.grid(row=1, column=0)
    button_fwd.grid(row=1, column=2)
    
    status = Label(root, text="Image "+ str(img_no) +" of " + str(len(img_list)), bd=1, relief=SUNKEN, anchor=E)
    status.grid(row=2, column=0, columnspan=3, sticky=W+E)
    
button_bck = Button(root, text="<<", command=back)
button_exit = Button(root, text="Exit", command=root.quit)
button_fwd = Button(root, text=">>", command=lambda: forward(2))

button_bck.grid(row=1, column=0)
button_exit.grid(row=1, column=1)
button_fwd.grid(row=1, column=2, pady=5)

root.mainloop()
```
